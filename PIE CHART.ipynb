```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
m=["andriod","ios","nokia","windows"]
s=[150,90,100,85 ]
color=['c','b','g','y']
explode=(0,0.2,0,0) # splitting the chart as cake pieces

plt.pie(s,labels=m,colors=color,shadow=False,
    startangle=90,explode=explode,autopct='%1.1f%%')
plt.title('mobile os market')
```
