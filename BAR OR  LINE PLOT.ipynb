# multi line chat for 2 lines

```python
product=pd.DataFrame({"year":[2014,2015,2016,2017,2018],
                    "prodAsales":[2000,25000,3000,32000,2200],
                    "prodBsales":[1200,2000,4500,3500,3000]})

ax=product.plot(x="year",y="prodAsales",kind="line",label="prodAsales",title="multi line plot",style="r--*")

product.plot(x="year",y="prodBsales",ax=ax,kind="line",label="prodBsales",title="multi line plot",style="b--*")
```

# multi line plot of 3 lines

```python
product=pd.DataFrame({"year":[2014,2015,2016,2017,2018],
                "prodAsales":[1000,5000,1300,3200,900],
                "prodBsales":[1200,2000,4500,3500,3000],
                "prodCsales":[3000,2300,3200,1200,2100]})


x=product.iloc[:,0] #year
a=product.iloc[:,1] #A
b=product.iloc[:,2] #B
c=product.iloc[:,3] #c
plt.plot(x,a,"-o",label="asales")
plt.plot(x,b,"--*",label="bsales")
plt.plot(x,c,"--+",label="csales")
plt.legend()
```

```python
df=pd.DataFrame({"year":[2020,2021,2022,2023],
                "covid":[15,200,100,50],
                "non-covids":[80,95,75,99],
                "fungi" :[50,60,70,80]})

ax=df.plot("year","covid",kind="line",label="covid",style="g-*")
df.plot("year","non-covids",kind="line",label="non-covid",style="r--*",ax=ax)
df.plot("year","fungi",kind="line",label="non-covid",style="b--*",ax=ax)
plt.xlabel("years")
plt.ylabel("no of people hospatalized with covid and non-covid")
```

# Multiple bar chat

```python
company=['infy','dell','google','wipro','tcs']
revenue=[123,345,654,122,899]
profit=[20,10,30,40,50]

plt.bar(company,revenue,label='revenue')
plt.bar(company,profit,label='profit')
plt.title("revenue/profit bar chat")
plt.legend()
```

```python
company=['infy','dell','google','wipro','tcs']
revenue=[123,345,654,122,899]
profit=[20,10,30,40,50]
xpos=np.arange(len(company))
#xpos=len(company)
#print(xpos)


plt.bar(xpos-0.2,revenue,width=0.4,label='revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='profit')
plt.title("revenue/profit bar chat")
plt.xlabel('company')
plt.ylabel('revenue')
plt.xticks(xpos,company) # gives the names of the company
plt.legend()
```
