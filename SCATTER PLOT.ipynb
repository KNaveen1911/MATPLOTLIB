```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.arange(0,10)  # x,y values must be same distribution
y=np.arange(11,21)
```

```python
a=np.arange(20,30)
b=np.arange(50,60)
```

# scatter plot

```python
plt.scatter(x,y,c='b') #color=blue
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2D graph')
plt.savefig('test.png')  # to save the image
```

```python
plt.plot(a,b,'r*--')
```

# sub plot

```python
plt.subplot(2,2,1) # 2 rows,2 columns,1st position 
plt.plot(x,y,'r--')
plt.subplot(2,2,2) # 2nd position
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
```

```python
x=np.arange(0,5*np.pi,1)
y=np.sin(x)
plt.plot(x,y)
```

```python
x=np.arange(40,50)
y=np.arange(50,60)

plt.scatter(x,y,c='r')
#plt.plot(x,y,'r*') 
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title("2d diagram")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r*-')
plt.subplot(2,2,2)
plt.plot(x,y,'go')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.subplot(2,2,4)
plt.plot(x,y,'g*-')
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

##subplot  of sine
plt.subplot(2,1,1)
plt.plot(x,y_sin,'g*')
plt.title('sine')
```

```python
##cosine
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r--')
plt.title('cosine')
plt.show()
```
